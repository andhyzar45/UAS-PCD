```python
pir=cv2.imread('pir.webp')
```

```python
pir=cv2.cvtColor(pir,cv2.COLOR_BGR2RGB)
hsv_pir=cv2.cvtColor(pir, cv2.COLOR_RGB2HSV)
```

```python
light_pir=(1,190,200)
dark_pir =(18, 255, 255)

```

```python
mask=cv2.inRange(hsv_pir,light_pir,dark_pir)
result=cv2.bitwise_and(pir,pir, mask=mask)
```

```python
plt.subplot(1,4,1)
plt.imshow(pir)
plt.subplot(1,4,2)
plt.imshow(hsv_pir)
plt.subplot(1,4,3)
plt.imshow(mask)
plt.subplot(1,4,4)
plt.imshow(result)
plt.show()
```
